```python
# Import library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from google.colab import files
uploaded = files.upload()
```

```python
# Membaca file CSV ke dalam DataFrame
df = pd.read_csv('CarPrice_Assignment.csv')
# Menampilkan lima baris pertama dari DataFrame
df.head
```

```python
# Mengecek duplikat
df.duplicated()
```

```python
# Menampilkan informasi DataFrame
df.info
```

```python
# Menampilkan ringkasan statistik
df.describe()
```

```python
# Memisahkan kolom-kolom berdasarkan jenis datanya
cat_col = [col for col in df.columns if df[col].dtype == 'object']
print('Categorical colums :',cat_col)
num_col = [col for col in df.columns if df[col].dtype != 'object']
print('Numberical columns:',num_col)
```

```python
# Menampilkan jumlah nilai unik dalam kolom-kolom kategori
df[cat_col].nunique()
```

```python
# Pisahkan fitur (features) dan label (target)
X = df.drop(['CarName', 'fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem'], axis=1)  # Sesuaikan dengan kolom target
y = df['horsepower']  # Sesuaikan dengan kolom target
```

```python
# Lakukan encoding untuk variabel kategorikal jika diperlukan
X_encoded = pd.get_dummies(X)
```

```python
# Pisahkan data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# Buat dan latih model Random Forest
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# Lakukan prediksi
y_pred = model.predict(X_test)
```

```python
# Evaluasi akurasi
accuracy = accuracy_score(y_test, y_pred)
print(f'Akurasi: {accuracy}')
```

```python
# Tampilkan laporan klasifikasi
print('Laporan Klasifikasi:\n', classification_report(y_test, y_pred))
```
